```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis_values = np.linspace(1,10,20)
x_axis_values
```

```python
y_axis_values = np.log(x_axis_values)
y_axis_values
```

```python
plt.plot(x_axis_values,y_axis_values,'blue')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title('Basic plot')
plt.show()
```

```python
#When are performing analysis we might want to place the plots side by side.
#The matplotlib's pyplot API provides us with the subplot option.
#For the subplot the first parameter is (nrows,ncolumns,index)

plt.subplot(121)
plt.plot(x_axis_values,y_axis_values,'blue')

plt.subplot(122)
plt.plot(x_axis_values,np.power(x_axis_values,2),'red')
```

<h3>Using the object oriented approach</h3>

<h4>When working with the pyplot API of the matplotlib it is recomended to use the object oriented approach
rather than the sequential one. The reason for using the ibject oriented approach is that it makes the work easir when there are multiple plots on the smae canvas and the library works in a much more flexible way. </h4>

```python
#using plt.subplots() is same as using fig = plt.figure() and then ax = fig.add_subplot(111)

fig,ax = plt.subplots()

ax.plot(x_axis_values,y_axis_values,'blue')
ax.set_xlabel('X_axis')
ax.set_ylabel('Y_axis')
ax.set_title('Basic plot')
```

```python
#WE can alo create two different subplots using the object oriented approach of the matplotlib
#When we are doing analysis we might sometimes want that we share the value of either y axis or the x axis and so
#for that we can do either sharex=True or shareY=True

fig,(ax1,ax2) = plt.subplots(1,2,sharey=True) #one row two columns
ax1.plot(x_axis_values,y_axis_values,'blue')
ax1.set_title('Sharing y axis')
ax2.scatter(x_axis_values,y_axis_values)
```

```python
#We might also want to plot two different fucntions on a same graph

fig,ax = plt.subplots()
ax.plot(x_axis_values,np.power(x_axis_values,2),label='Square')#We can also provide labels to associate with fucntions
ax.plot(y_axis_values,np.power(x_axis_values,3),label='Cube')
ax.legend() #To create the box which tells us color which line belongs to which function 
```

```python
#WE can also create a scatter plot.
#scatter plot is generally used to find out the relationship between the x and y values

fig,ax = plt.subplots()
ax.scatter(x_axis_values,np.power(x_axis_values,3),label='cube')
ax.scatter(x_axis_values,np.power(x_axis_values,4),label='power 4')
ax.scatter(x_axis_values,np.power(x_axis_values,5),label='power 5')
ax.scatter(x_axis_values,np.power(x_axis_values,6),label='power 6')

ax.legend()
#As we increase the power the number gets a greater difference to its previous number
#Eg 10^2 = 100, 9^2=81(difference = 19)
#   10^3 = 1000, 9^3=729(difference = 271)
#   10^4 = 10000, 9^4=6561(difference=5561) etc..
```
